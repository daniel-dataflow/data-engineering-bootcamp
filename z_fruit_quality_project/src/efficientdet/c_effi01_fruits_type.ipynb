```python
# 1. í™˜ê²½ ì„¤ì • ë° ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
import os
import json
import glob
import torch
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

print(f"PyTorch version: {torch.__version__}")
print(f"CUDA available: {torch.cuda.is_available()}")

# ê²½ë¡œ ì„¤ì •
BASE_DIR = os.path.abspath('../../data')
DATASET_DIR = os.path.join(BASE_DIR, 'row')
IMAGES_DIR = os.path.join(DATASET_DIR, 'images')
LABELS_DIR = os.path.join(DATASET_DIR, 'json_labels')
OUTPUT_DIR = os.path.join(BASE_DIR, '../../data/efficientdet', 'outputs')
os.makedirs(OUTPUT_DIR, exist_ok=True)

print(f"Base Directory: {BASE_DIR}")
print(f"Dataset Directory: {DATASET_DIR}")
print(f"Images Directory: {IMAGES_DIR}")
print(f"Labels Directory: {LABELS_DIR}")
```

```python
# 2. ë°ì´í„°ì…‹ ë¶„ì„ ë° í†µê³„
def analyze_dataset():
    """ë°ì´í„°ì…‹ êµ¬ì¡° ë¶„ì„"""
    json_files = glob.glob(os.path.join(LABELS_DIR, '*.json'))
    
    class_counts = {'ì‚¬ê³¼': 0, 'ë°°': 0, 'ê°': 0}
    total_boxes = 0
    
    print(f"\nì´ JSON íŒŒì¼ ìˆ˜: {len(json_files)}")
    
    for json_file in json_files[:5]:  # ìƒ˜í”Œ 5ê°œ ì¶œë ¥
        with open(json_file, 'r', encoding='utf-8') as f:
            data = json.load(f)
            print(f"\níŒŒì¼: {os.path.basename(json_file)}")
            print(f"  cate1: {data.get('cate1', 'N/A')}")
            print(f"  cate3: {data.get('cate3', 'N/A')}")
            print(f"  bndbox: {data.get('bndbox', {})}")
    
    # ì „ì²´ í†µê³„
    for json_file in json_files:
        with open(json_file, 'r', encoding='utf-8') as f:
            data = json.load(f)
            cate1 = data.get('cate1', '')
            if cate1 in class_counts:
                class_counts[cate1] += 1
                total_boxes += 1
    
    print(f"\n=== ë°ì´í„°ì…‹ í†µê³„ ===")
    print(f"ì´ ì´ë¯¸ì§€ ìˆ˜: {len(json_files)}")
    print(f"ì´ ê°ì²´ ìˆ˜: {total_boxes}")
    for cls, count in class_counts.items():
        print(f"  {cls}: {count}ê°œ ({count/total_boxes*100:.1f}%)")
    
    return class_counts

stats = analyze_dataset()
```

```python
# 3. COCO í˜•ì‹ ë°ì´í„°ì…‹ ë³€í™˜
class FruitDatasetConverter:
    """JSON to COCO format converter for cate1 (fruit type)"""
    
    def __init__(self, images_dir, labels_dir, output_dir):
        self.images_dir = images_dir
        self.labels_dir = labels_dir
        self.output_dir = output_dir
        
        # cate1 ê¸°ë°˜ í´ë˜ìŠ¤ ë§¤í•‘
        self.class_map = {
            'ì‚¬ê³¼': 0,
            'ë°°': 1,
            'ê°': 2
        }
        self.classes = ['ì‚¬ê³¼', 'ë°°', 'ê°']
        
    def convert_to_coco(self, split_ratio=0.8):
        """COCO í˜•ì‹ìœ¼ë¡œ ë³€í™˜ ë° train/val ë¶„í• """
        
        json_files = sorted(glob.glob(os.path.join(self.labels_dir, '*.json')))
        np.random.seed(42)
        np.random.shuffle(json_files)
        
        split_idx = int(len(json_files) * split_ratio)
        train_files = json_files[:split_idx]
        val_files = json_files[split_idx:]
        
        print(f"\në°ì´í„° ë¶„í• :")
        print(f"  Train: {len(train_files)}ê°œ")
        print(f"  Val: {len(val_files)}ê°œ")
        
        # Train/Val ê°ê° ë³€í™˜
        train_coco = self._create_coco_dataset(train_files, 'train')
        val_coco = self._create_coco_dataset(val_files, 'val')
        
        # ì €ì¥
        train_json_path = os.path.join(self.output_dir, 'train_cate1.json')
        val_json_path = os.path.join(self.output_dir, 'val_cate1.json')
        
        with open(train_json_path, 'w', encoding='utf-8') as f:
            json.dump(train_coco, f, ensure_ascii=False, indent=2)
        
        with open(val_json_path, 'w', encoding='utf-8') as f:
            json.dump(val_coco, f, ensure_ascii=False, indent=2)
        
        print(f"\nâœ… COCO íŒŒì¼ ìƒì„± ì™„ë£Œ:")
        print(f"  Train: {train_json_path}")
        print(f"  Val: {val_json_path}")
        
        return train_json_path, val_json_path
    
    def _create_coco_dataset(self, json_files, split_name):
        """COCO ë°ì´í„°ì…‹ êµ¬ì¡° ìƒì„±"""
        
        coco_dataset = {
            "images": [],
            "annotations": [],
            "categories": []
        }
        
        # Categories ì¶”ê°€
        for idx, class_name in enumerate(self.classes):
            coco_dataset["categories"].append({
                "id": idx,
                "name": class_name,
                "supercategory": "fruit"
            })
        
        annotation_id = 0
        
        for image_id, json_file in enumerate(tqdm(json_files, desc=f"Converting {split_name}")):
            with open(json_file, 'r', encoding='utf-8') as f:
                data = json.load(f)
            
            # ì´ë¯¸ì§€ íŒŒì¼ëª…
            img_filename = os.path.basename(json_file).replace('.json', '.png')
            img_path = os.path.join(self.images_dir, img_filename)
            
            if not os.path.exists(img_path):
                continue
            
            # ì´ë¯¸ì§€ ì •ë³´
            img = Image.open(img_path)
            width, height = img.size
            
            coco_dataset["images"].append({
                "id": image_id,
                "file_name": img_filename,
                "width": width,
                "height": height
            })
            
            # Annotation ì¶”ê°€
            cate1 = data.get('cate1', '')
            if cate1 not in self.class_map:
                continue
            
            bbox = data.get('bndbox', {})
            xmin = float(bbox.get('xmin', 0))
            ymin = float(bbox.get('ymin', 0))
            xmax = float(bbox.get('xmax', width))
            ymax = float(bbox.get('ymax', height))
            
            bbox_width = xmax - xmin
            bbox_height = ymax - ymin
            area = bbox_width * bbox_height
            
            coco_dataset["annotations"].append({
                "id": annotation_id,
                "image_id": image_id,
                "category_id": self.class_map[cate1],
                "bbox": [xmin, ymin, bbox_width, bbox_height],  # COCO format: [x, y, w, h]
                "area": area,
                "iscrowd": 0
            })
            
            annotation_id += 1
        
        print(f"{split_name} - Images: {len(coco_dataset['images'])}, Annotations: {len(coco_dataset['annotations'])}")
        
        return coco_dataset

# ë°ì´í„°ì…‹ ë³€í™˜ ì‹¤í–‰
converter = FruitDatasetConverter(IMAGES_DIR, LABELS_DIR, OUTPUT_DIR)
train_json, val_json = converter.convert_to_coco(split_ratio=0.8)
```

```python
# 4. PyTorch Dataset í´ë˜ìŠ¤
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms

class COCOFruitDataset(Dataset):
    """COCO í˜•ì‹ ê³¼ì¼ ë°ì´í„°ì…‹"""
    
    def __init__(self, images_dir, coco_json_path, transform=None):
        self.images_dir = images_dir
        self.transform = transform
        
        with open(coco_json_path, 'r') as f:
            self.coco_data = json.load(f)
        
        self.images = self.coco_data['images']
        self.annotations = self.coco_data['annotations']
        
        # image_idë¥¼ í‚¤ë¡œ í•˜ëŠ” annotations ë”•ì…”ë„ˆë¦¬ ìƒì„±
        self.img_to_anns = {}
        for ann in self.annotations:
            img_id = ann['image_id']
            if img_id not in self.img_to_anns:
                self.img_to_anns[img_id] = []
            self.img_to_anns[img_id].append(ann)
    
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        img_info = self.images[idx]
        img_path = os.path.join(self.images_dir, img_info['file_name'])
        
        # ì´ë¯¸ì§€ ë¡œë“œ
        image = Image.open(img_path).convert('RGB')
        image = np.array(image)
        
        # Annotations ê°€ì ¸ì˜¤ê¸°
        img_id = img_info['id']
        anns = self.img_to_anns.get(img_id, [])
        
        boxes = []
        labels = []
        
        for ann in anns:
            x, y, w, h = ann['bbox']
            boxes.append([x, y, x + w, y + h])  # [xmin, ymin, xmax, ymax]
            labels.append(ann['category_id'])
        
        boxes = np.array(boxes, dtype=np.float32)
        labels = np.array(labels, dtype=np.int64)
        
        target = {
            'boxes': boxes,
            'labels': labels,
            'image_id': img_id
        }
        
        if self.transform:
            image = self.transform(image)
        
        return image, target

# Transform ì •ì˜
transform = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((512, 512)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# ë°ì´í„°ì…‹ ë° DataLoader ìƒì„±
train_dataset = COCOFruitDataset(IMAGES_DIR, train_json, transform=transform)
val_dataset = COCOFruitDataset(IMAGES_DIR, val_json, transform=transform)

print(f"\nâœ… ë°ì´í„°ì…‹ ë¡œë“œ ì™„ë£Œ:")
print(f"  Train ìƒ˜í”Œ ìˆ˜: {len(train_dataset)}")
print(f"  Val ìƒ˜í”Œ ìˆ˜: {len(val_dataset)}")

# ìƒ˜í”Œ ë°ì´í„° í™•ì¸
sample_img, sample_target = train_dataset[0]
print(f"\nìƒ˜í”Œ ë°ì´í„°:")
print(f"  Image shape: {sample_img.shape}")
print(f"  Boxes: {sample_target['boxes']}")
print(f"  Labels: {sample_target['labels']}")
```

```python
# 5. EfficientDet ëª¨ë¸ ë¡œë“œ (effdet ë¼ì´ë¸ŒëŸ¬ë¦¬ ì‚¬ìš©)
# effdet ì„¤ì¹˜: !pip install effdet

try:
    from effdet import get_efficientdet_config, EfficientDet, DetBenchTrain
    from effdet.efficientdet import HeadNet
except ImportError:
    print("âš ï¸  effdet ë¼ì´ë¸ŒëŸ¬ë¦¬ê°€ ì„¤ì¹˜ë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")
    print("ë‹¤ìŒ ëª…ë ¹ì–´ë¡œ ì„¤ì¹˜í•˜ì„¸ìš”:")
    print("!pip install effdet")
    print("ë˜ëŠ” conda í™˜ê²½ì—ì„œ:")
    print("pip install effdet")

# EfficientDet-D1 ì„¤ì •
def create_model(num_classes=3):
    """EfficientDet-D1 ëª¨ë¸ ìƒì„±"""
    
    config = get_efficientdet_config('tf_efficientdet_d1')
    config.num_classes = num_classes
    config.image_size = (512, 512)
    
    net = EfficientDet(config, pretrained_backbone=True)
    net.class_net = HeadNet(
        config,
        num_outputs=config.num_classes,
    )
    
    model = DetBenchTrain(net, config)
    
    return model, config

model, config = create_model(num_classes=3)
device = torch.device('cpu')
model = model.to(device)

print(f"\nâœ… EfficientDet-D1 ëª¨ë¸ ë¡œë“œ ì™„ë£Œ")
print(f"  Device: {device}")
print(f"  Input size: {config.image_size}")
print(f"  Classes: {config.num_classes}")
```

```python
# 6. í•™ìŠµ í•¨ìˆ˜ ì •ì˜
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR

def collate_fn(batch):
    """ë°°ì¹˜ ë°ì´í„° collate í•¨ìˆ˜"""
    images = []
    targets = []
    
    for img, target in batch:
        images.append(img)
        targets.append(target)
    
    images = torch.stack(images, dim=0)
    
    return images, targets

def train_one_epoch(model, dataloader, optimizer, device, epoch):
    """1 ì—í­ í•™ìŠµ"""
    model.train()
    
    total_loss = 0
    num_batches = 0  # ì´ ì¤„ ì¶”ê°€
    progress_bar = tqdm(dataloader, desc=f"Epoch {epoch+1}")
    
    for images, targets in progress_bar:
        images = images.to(device)
        
        # EfficientDetì€ targetsë¥¼ íŠ¹ë³„í•œ í˜•ì‹ìœ¼ë¡œ ë°›ìŒ
        batch_size = images.size(0)
        
        # targetsë¥¼ ëª¨ë¸ì´ ìš”êµ¬í•˜ëŠ” í˜•ì‹ìœ¼ë¡œ ë³€í™˜
        target_dict = {
            'bbox': [],
            'cls': []
        }
        
        for target in targets:
            boxes = torch.tensor(target['boxes'], dtype=torch.float32)
            labels = torch.tensor(target['labels'], dtype=torch.int64)
            
            target_dict['bbox'].append(boxes)
            target_dict['cls'].append(labels)
        
        # Forward
        optimizer.zero_grad()
        
        try:
            output = model(images, target_dict)
            
            # outputì´ íŠœí”Œì¸ ê²½ìš° ì²« ë²ˆì§¸ ìš”ì†Œê°€ loss
            if isinstance(output, tuple):
                loss = output[0]
            else:
                loss = output
            
            # lossê°€ dictì¸ ê²½ìš°
            if isinstance(loss, dict):
                loss = sum(loss.values())
            
            # lossê°€ í…ì„œì¸ì§€ í™•ì¸
            if isinstance(loss, torch.Tensor):
                if loss.dim() > 0:
                    loss = loss.mean()
            else:
                print(f"\nâš ï¸  ì˜ˆìƒì¹˜ ëª»í•œ loss íƒ€ì…: {type(loss)}")
                continue
            
            # Backward
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            num_batches += 1
            progress_bar.set_postfix({'loss': loss.item()})

        except Exception as e:
            print(f"\nâš ï¸  ë°°ì¹˜ ì²˜ë¦¬ ì¤‘ ì—ëŸ¬ ë°œìƒ: {e}")
            import traceback
            traceback.print_exc()
            continue
    
    avg_loss = total_loss / max(num_batches, 1)
    return avg_loss

# DataLoader ìƒì„±
train_loader = DataLoader(
    train_dataset,
    batch_size=4,  # CPUì—ì„œëŠ” ì‘ì€ ë°°ì¹˜ ì‚¬ìš©
    shuffle=True,
    collate_fn=collate_fn,
    num_workers=0
)

val_loader = DataLoader(
    val_dataset,
    batch_size=4,
    shuffle=False,
    collate_fn=collate_fn,
    num_workers=0
)

print(f"\nâœ… DataLoader ìƒì„± ì™„ë£Œ")
print(f"  Train batches: {len(train_loader)}")
print(f"  Val batches: {len(val_loader)}")
```

```python
# 7. í•™ìŠµ ì‹¤í–‰
NUM_EPOCHS = 5
LEARNING_RATE = 1e-4

optimizer = AdamW(model.parameters(), lr=LEARNING_RATE, weight_decay=1e-4)
scheduler = CosineAnnealingLR(optimizer, T_max=NUM_EPOCHS, eta_min=1e-6)

print(f"\nğŸš€ í•™ìŠµ ì‹œì‘...")
print(f"  Epochs: {NUM_EPOCHS}")
print(f"  Learning Rate: {LEARNING_RATE}")
print(f"  Optimizer: AdamW")
print(f"  Scheduler: CosineAnnealingLR")
print(f"\n{'='*60}\n")

train_losses = []
best_loss = float('inf')

for epoch in range(NUM_EPOCHS):
    # í•™ìŠµ
    train_loss = train_one_epoch(model, train_loader, optimizer, device, epoch)
    train_losses.append(train_loss)
    
    # Learning rate ì—…ë°ì´íŠ¸
    scheduler.step()
    
    print(f"Epoch {epoch+1}/{NUM_EPOCHS} - Loss: {train_loss:.4f} - LR: {scheduler.get_last_lr()[0]:.6f}")
    
    # ëª¨ë¸ ì €ì¥ (Best)
    if train_loss < best_loss:
        best_loss = train_loss
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': train_loss,
        }, os.path.join(OUTPUT_DIR, 'best_efficientdet_cate1.pth'))
        print(f"  âœ… Best ëª¨ë¸ ì €ì¥ (Loss: {best_loss:.4f})")
    
    # ì£¼ê¸°ì  ì €ì¥ (10 epochë§ˆë‹¤)
    if (epoch + 1) % 10 == 0:
        torch.save({
            'epoch': epoch,
            'model_state_dict': model.state_dict(),
            'optimizer_state_dict': optimizer.state_dict(),
            'loss': train_loss,
        }, os.path.join(OUTPUT_DIR, f'efficientdet_cate1_epoch{epoch+1}.pth'))
        print(f"  ğŸ’¾ Checkpoint ì €ì¥ (Epoch {epoch+1})")
    
    print()

print(f"\nğŸ‰ í•™ìŠµ ì™„ë£Œ!")
print(f"  Best Loss: {best_loss:.4f}")
print(f"  ëª¨ë¸ ì €ì¥ ê²½ë¡œ: {OUTPUT_DIR}")
```

```python
# 9. ì¶”ë¡  ë° ì‹œê°í™”

import platform
# ê¸€ì”¨ì²´ ì§€ì •
if platform.system() == 'Darwin': # macOS
    plt.rcParams['font.family'] = 'AppleGothic'
elif platform.system() == 'Windows': # Windows
    plt.rcParams['font.family'] = 'Malgun Gothic'

# (ë³´ë„ˆìŠ¤) ë§ˆì´ë„ˆìŠ¤ ë¶€í˜¸ ê¹¨ì§ ë°©ì§€
plt.rcParams['axes.unicode_minus'] = False

def visualize_predictions(model, dataset, device, num_samples=6):
    """ì˜ˆì¸¡ ê²°ê³¼ ì‹œê°í™”"""
    # ì¶”ë¡ ìš© ëª¨ë¸ë¡œ ë³€ê²½
    model.eval()
    
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.flatten()
    
    class_names = ['ì‚¬ê³¼', 'ë°°', 'ê°']
    colors = ['red', 'green', 'blue']
    
    sample_idx = 0
    
    for i in range(len(dataset)):
        if sample_idx >= num_samples:
            break
            
        img, target = dataset[i]
        
        # ë¹ˆ ë°•ìŠ¤ ê±´ë„ˆë›°ê¸°
        if len(target['boxes']) == 0:
            continue
        
        # ì´ë¯¸ì§€ denormalize
        img_np = img.cpu().numpy().transpose(1, 2, 0)
        img_np = img_np * np.array([0.229, 0.224, 0.225]) + np.array([0.485, 0.456, 0.406])
        img_np = np.clip(img_np, 0, 1)
        
        axes[sample_idx].imshow(img_np)
        axes[sample_idx].axis('off')
        axes[sample_idx].set_title(f'Sample {sample_idx+1}', fontsize=10, fontweight='bold')
        
        # Ground Truth ë°•ìŠ¤ë§Œ í‘œì‹œ (ì¶”ë¡  ê²°ê³¼ëŠ” ìƒëµ)
        gt_boxes = target['boxes']
        gt_labels = target['labels']
        
        for box, label in zip(gt_boxes, gt_labels):
            x1, y1, x2, y2 = box
            # ì •ê·œí™”ëœ ì¢Œí‘œë¥¼ ì´ë¯¸ì§€ í¬ê¸°ë¡œ ë³€í™˜
            width, height = 512, 512
            x1, x2 = x1 * width / 1000, x2 * width / 1000
            y1, y2 = y1 * height / 1000, y2 * height / 1000
            
            rect = plt.Rectangle(
                (x1, y1), x2 - x1, y2 - y1,
                fill=False, color=colors[int(label)], linewidth=2
            )
            axes[sample_idx].add_patch(rect)
            axes[sample_idx].text(
                x1, y1 - 5,
                f'{class_names[int(label)]}',
                color=colors[int(label)],
                fontsize=9,
                fontweight='bold',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.7)
            )
        
        sample_idx += 1
    
    # ë‚¨ì€ subplot ìˆ¨ê¸°ê¸°
    for j in range(sample_idx, 6):
        axes[j].axis('off')
    
    plt.tight_layout()
    viz_path = os.path.join(OUTPUT_DIR, 'predictions_cate1.png')
    plt.savefig(viz_path, dpi=150, bbox_inches='tight')
    plt.show()
    
    print(f"\nâœ… ì˜ˆì¸¡ ì‹œê°í™” ì €ì¥: {viz_path}")
    print("(Ground Truth ë°•ìŠ¤ë§Œ í‘œì‹œë©ë‹ˆë‹¤)")

# ì‹œê°í™” ì‹¤í–‰
visualize_predictions(model, val_dataset, device, num_samples=6)
```
