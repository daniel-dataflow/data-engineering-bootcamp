```python
import os
from pathlib import Path
from tqdm import tqdm

BASE_DIR = Path.cwd()  # í˜„ì¬ ì‹¤í–‰ í´ë”
IMG_DIR = BASE_DIR / "datasets/images"
JSON_DIR = BASE_DIR / "datasets/json_labels"


def clean_unmatched_files(delete=False):
    img_files = {p.stem for p in IMG_DIR.glob("*.png")}
    json_files = {p.stem for p in JSON_DIR.glob("*.json")}

    # --- ë§¤ì¹­ë˜ì§€ ì•Šì€ íŒŒì¼ íƒìƒ‰ ---
    imgs_without_json = img_files - json_files
    jsons_without_img = json_files - img_files

    print("\n==============================")
    print("ğŸ“Š ë§¤ì¹­ ê²€ì‚¬ ê²°ê³¼")
    print("==============================")
    print(f"ì´ ì´ë¯¸ì§€ ìˆ˜: {len(img_files)}")
    print(f"ì´ JSON ìˆ˜: {len(json_files)}")
    print(f"ì´ë¯¸ì§€ë§Œ ì¡´ì¬í•˜ëŠ” íŒŒì¼: {len(imgs_without_json)}ê°œ")
    print(f"JSONë§Œ ì¡´ì¬í•˜ëŠ” íŒŒì¼: {len(jsons_without_img)}ê°œ")
    print("==============================\n")

    # --- ì‚­ì œ ë˜ëŠ” ë¯¸ë¦¬ë³´ê¸° ---
    if imgs_without_json:
        print("ğŸ–¼ï¸ JSONì´ ì—†ëŠ” ì´ë¯¸ì§€ íŒŒì¼ ëª©ë¡:")
        for name in sorted(imgs_without_json):
            print(f" - {name}.png")
        if delete:
            print("\nğŸ—‘ï¸ JSONì´ ì—†ëŠ” ì´ë¯¸ì§€ íŒŒì¼ ì‚­ì œ ì¤‘...")
            for name in tqdm(imgs_without_json):
                (IMG_DIR / f"{name}.png").unlink(missing_ok=True)

    if jsons_without_img:
        print("\nğŸ“„ ì´ë¯¸ì§€ê°€ ì—†ëŠ” JSON íŒŒì¼ ëª©ë¡:")
        for name in sorted(jsons_without_img):
            print(f" - {name}.json")
        if delete:
            print("\nğŸ—‘ï¸ ì´ë¯¸ì§€ê°€ ì—†ëŠ” JSON íŒŒì¼ ì‚­ì œ ì¤‘...")
            for name in tqdm(jsons_without_img):
                (JSON_DIR / f"{name}.json").unlink(missing_ok=True)

    if not imgs_without_json and not jsons_without_img:
        print("âœ… ëª¨ë“  íŒŒì¼ì´ ì •ìƒì ìœ¼ë¡œ ë§¤ì¹­ë˜ì—ˆìŠµë‹ˆë‹¤!")

    print("\n-------------------------------------")
    print("ğŸ’¡ ì‹¤ì œ ì‚­ì œë¥¼ ì›í•  ê²½ìš°, í•¨ìˆ˜ í˜¸ì¶œ ì‹œ delete=True ë¡œ ë³€ê²½í•˜ì„¸ìš”.")
    print("ì˜ˆ: clean_unmatched_files(delete=True)")
    print("-------------------------------------")

if __name__ == "__main__":
    clean_unmatched_files(delete=True)
```
