```python
import torch
import os
import cv2 # OpenCV 

# 1.  
# PyTorch Hub  'ultralytics/yolov5'  
# 'yolov5s' (small)  .
# (    1 )
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# 2.   
# test.ipynb  'fruits'   .
# 'data'  'fruits'   ,  'data/.jpg' .
img_path = 'data/image/apple_fuji_M_1-1_1TOP.png'  # <---      .

# 3.   
results = model(img_path)

# 4.   (Jupyter Notebook  )
#       .
results.show()

# 5. ()   
# (print)     .
results.print()

# 6. ()   Pandas  
#   (xyxy), (confidence),  ID, (name)    .
print(results.pandas().xyxy[0])
```
