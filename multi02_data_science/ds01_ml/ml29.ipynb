```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from xgboost import plot_importance

import matplotlib.pyplot as plt
```

```python
cencer = load_breast_cancer()

x_train, x_test, y_train, y_test = train_test_split(cencer.data, cencer.target, test_size=0.2, random_state=1)
```

```python
# datasets
dtrain = xgb.DMatrix(data=x_train, label=y_train)
dtest = xgb.DMatrix(data=x_test, label=y_test)
```

```python
"""
max_depth : tree 의 최대 깊이
eta : learning_rate
objective : 분류 설정
    reg:linear -> 회귀
    binary:logistic -> 이진분류
    multi:softmax -> 다항분류(클래스)
    multi:softprob : 다항분류 (확률)
eval_metric : 손실함수 (클래스 분류 - error / 다른 것들은?)
"""

params = {
    "max_depth" : 3,
    "eta" : 0.1,
    "objective" : "binary:logistic",
    "eval_metric" : "error"
}

num_boost_round = 400
```

```python
# training / evaluation (valid, test)
data_list =[(dtrain, "train"), (dtest, "eval")]

xgb_model = xgb.train(params=params, dtrain=dtrain, num_boost_round=num_boost_round, evals=data_list)
```

```python
predict = xgb_model.predict(dtest)
predict
```

```python
predict_list = [1 if x > 0.5 else 0 for x in predict]
predict_list
```

```python
accuracy_score(y_test, predict_list)
```

```python
fig, ax = plt.subplots(figsize=(10, 12))
plot_importance(xgb_model, ax=ax)
```
