```python
import pandas as pd

from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
corpus = ["A hundred bad days made a hundred good stories",
          "A hundred good stories make me interesting at parties"]
```

```python
cv = CountVectorizer(stop_words=stopwords.words("english"))
cv.fit(corpus)
bow = cv.transform(corpus)
df = pd.DataFrame(bow.toarray(), index=corpus, columns=cv.get_feature_names_out())
df
```

```python
"""
cosine similarity : vector 간의 cosine 각도를 이용하여 유사도 분석
    각도가 0 으로 가까울수록 cosine 값은 1에 가깝다 =  유사하다.
    각도가 90 으로 가까울수록 cosine 값은 0에 가깝다 =  유사하지 않다.
"""

cosine_similarity(bow[0], bow[1])
```
