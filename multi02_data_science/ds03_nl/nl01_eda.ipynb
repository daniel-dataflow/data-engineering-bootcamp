NLP (Natural Language Processing) : 자연어 처리

```python
with open("../datasets/shakespeare.txt") as f:
    shakespeare = f.readlines()

shakespeare
```

```python
# 공백으로 거대한 문장으로 만들었다.
tokens = " ".join(shakespeare)
```

```python
# 소문자로 변경
tokens = tokens.lower()
```

```python
# \n  제거
tokens = tokens.replace("\n", "")
tokens
```

```python
import re
```

```python
# a부터 z까지 아닌 것들을 공백으로 변경해줘.
tokens = re.sub("[^a-z]", " ", tokens)
tokens
```

```python
tokens =tokens.split(" ")
```

```python
tokens
```

```python
# '', 삭제
tokens = list(filter(None, tokens))
tokens
```

```python
from collections import Counter
```

```python
# 어떤 단어가 몇개가 있다라고 보여줌
tokens_cnt = Counter(tokens)
tokens_cnt
```

```python
# 딕셔너리로 변경, 순서는 랜덤으로
dict(tokens_cnt)
```

```python
import pandas as pd
```

```python
# 데이터를 사용하기 위해서 데이터 프레임으로 만들어서 사용할 예정이다.
token_df = pd.DataFrame.from_dict(tokens_cnt, orient="index", columns=["freq"])
token_df.index.name = "token"
token_df
```

```python
# 빈도수 높은 순으로 정렬
token_df.sort_values(by="freq", ascending=False, inplace=True)
token_df
```

```python
# 그림으로 그려보자.
import matplotlib.pyplot as plt
```

```python
ax = token_df.head(20).plot(kind="barh")
ax.invert_yaxis()
```
