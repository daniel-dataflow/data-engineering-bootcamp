```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.3, random_state=1)
```

```python
print(f"train : {x_train.shape} / {y_train.shape}")
print(f"valid : {x_val.shape} / {y_val.shape}")
print(f"test : {x_test.shape} / {y_test.shape}")
```

```python
# 모델한테 주기 위해서 3차원을 2차원으로 다시 재구성
# 배열을 이어붙이기 형식으로 2차원으로 변경하는 형태로 만들어 주었다.
# 데이터가 변경되는 것 같을지 모르지만 원래의 값을 학습을 시켜줬기 때문에 그것을 추론할 수 있다.
x_train_scale = (x_train / 255.0).reshape(x_train.shape[0], 28 * 28)
x_val_scale = (x_val / 255.0).reshape(x_val.shape[0], 28 * 28)
x_test_scale = (x_test / 255.0).reshape(x_test.shape[0], 28 * 28)
```

```python
print(x_train_scale.shape)
print(x_val_scale.shape)
print(x_test_scale.shape)
```

```python
# 분류를 하고 싶다.
# y는 0부터 9까지 있는데 순차적인 데이터이다 원핫 인코딩으로 하자
# 소프트 맥스는 확률을 계산하는데 범주형 데이터로 계산한다.
# 10개로 분류할거야
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train_cate = to_categorical(y_train)
y_val_cate = to_categorical(y_val)
y_test_cate = to_categorical(y_test)
```

```python
print(y_train_cate.shape)
print(y_val_cate.shape)
print(y_test_cate.shape)
```

```python
print(y_train[0])
print(y_train_cate[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
# input layer 784는 28*28의 문자가 길게 이어져서 나오는 값이다.
model.add(Dense(8, input_shape=(784, ), activation="relu"))

# hidden layer
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))

# output layer
model.add(Dense(10, activation="softmax"))
```

```python
model.summary()
# 어떤 레이어들이 있는지 보여주고 있다.
# param : w와 b의 갯수
# input layer param : 786 (x) * 8 (W) + 8 (b)
# 
# 퍼세트론당 1개가 들어가야 하기 때문에 8개이다.
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model, show_shapes=True)
```

```python
# categorical_crossentropy
# cross entropy : 예측값이 실제값과 다를 확률을 계산 (분류 모델)
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["acc"])
```

```python
result = model.fit(x_train_scale, y_train_cate, epochs=100, batch_size=100, validation_data=(x_val_scale, y_val_cate))
```

```python
result_dict = result.history
result_dict.keys()
```

```python
loss = result_dict["loss"]
val_loss = result_dict["val_loss"]

acc = result_dict["acc"]
val_acc = result_dict["val_acc"]
```

```python
epochs = range(1, len(loss) + 1)
```

```python
# 파란색이 내려가는것은 학습을 잘하고 있다.
# 빨간색이 올라가는 것은 과적합의 현상이 나타나고 있다.
# 빨간색이 올라가기 전의 파란색의 구간이 가장 학습을 잘하고 있다고 판단 할 수 있다.
plt.plot(epochs, loss, color="blue", label="loss")
plt.plot(epochs, val_loss, color="red", label="val_loss")

plt.xlabel("epochs")
plt.ylabel("loss")
plt.title("loss")
plt.legend()
plt.show()
```

```python
# 정확도
# 파란색이 점점 잘맞추고 있다.
# 검증데이터 빨간색은 파란색과 비슷하게 흘러간다. 과적합으로 우려된다.
plt.plot(epochs, acc, color="blue", label="acc")
plt.plot(epochs, val_acc, color="red", label="val_acc")

plt.xlabel("epochs")
plt.ylabel("acc")
plt.title("acc")
plt.legend()
plt.show()
```

```python
# 평가하자
# loss는 26% 어큐러시 : 94%이상으로 맞추고 있다.
model.evaluate(x_test_scale, y_test_cate)
```
