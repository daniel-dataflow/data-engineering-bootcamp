```python
import numpy as np

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse
```

```python
# 분류문제입니다.
# xor
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
```

```python
# layer 를 순차적으로 연결
model = Sequential()
```

```python
# Dense : layer, 출력이 1개이기 때문에 퍼세트론 1개, 분류이기 때문에 linear -> sigmoid(0.5 기준으로 크거나 작다)로 바꾼다.
model.add(Dense(1, input_shape=(2, ), activation="sigmoid"))
```

```python
# SGD : Stochastic Gradient Descent (확률적 경사하강법)
# metrics : evaluate (평가) 시 추가할 항목 -> 추가하지 않으면 loss값 만 나온다!
model.compile(loss=mse, optimizer=SGD(), metrics=["acc"])
```

```python
# epochs : 학습횟수
model.fit(x, y, epochs=1000)
```

```python
# evaluate : 평가 -> loss [acc]

model.evaluate(x, y)
```

```python
# 분류해줘
predict = model.predict(x)
predict
```

```python
# 너가 계산한 값보여줘
# 첫번째에는 W1(weight) W2 가 두번째에는 b(바이어스)
model.get_weights()
```
