```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../images/headache.png")
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
# 행 : 높이 / 열 : 폭 / 채널 : 색상
img.shape
```

```python
# 색상 : BGR (opencv) <-> RGB (matplotlib)
# Blue, Green, Red
b, g, r = cv.split(img)
img_rgb = cv.merge([r, g, b])
plt.imshow(img_rgb)
```

```python
img_rgb2 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb2)
```
