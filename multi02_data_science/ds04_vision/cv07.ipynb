```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

HSV

- H : Hue (색상) 0 ~ 360 -> opencv 에선 0 ~ 179 (2로 나눈 값 -> 8bit 255까지 밖에 표현 못함)

    - red : 0
    - yellow : 60
    - green : 120
    - cyan : 180
    - blue : 240
    - magenta : 300

- S : Saturation (채도) 100% -> opencv 에선 0 ~ 255

- V : Value (명도) 100% -> 0 ~ 255

```python
img = cv.imread("../images/roselia.png")

img_plot = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plot)
```

```python
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
img_hsv.shape
```

```python
"""
cv.imshow("hsv", img_hsv)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(img_hsv, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
# blue : 240 -> opencv 에서는 120
# 로우블루와 어퍼블루 사이에 있는 값(100-140)은 하얀색 그외의 값은 검은색으로 바꿔라.
lower_blue = (120 - 20, 0, 0)
upper_blue = (120 + 20, 255, 255)

img_blue_mask = cv.inRange(img_hsv, lower_blue, upper_blue)
```

```python
"""
cv.imshow("blue", img_img_blue_maskmedian)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(img_blue_mask, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
result_blue = cv.bitwise_and(img, img, mask=img_blue_mask)
```

```python
"""
cv.imshow("blue", result_blue)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(result_blue, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
img01 = cv.imread("../images/headache.png")
img02 = cv.imread("../images/pika.png")
```

```python
img_and = cv.bitwise_and(img01, img02)
img_or = cv.bitwise_or(img01, img02)
img_xor = cv.bitwise_xor(img01, img02)
img_not = cv.bitwise_not(img01, img02)
```

```python
"""
cv.imshow("and", img_and)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(img_and, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
"""
cv.imshow("or", img_or)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(img_or, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
"""
cv.imshow("xor", img_xor)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(img_xor, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
"""
cv.imshow("not", img_not)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(img_not, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```
