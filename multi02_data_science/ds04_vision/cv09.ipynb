```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../images/headache.png")

img_plot = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plot)
```

```python
h, w, c = img.shape
mat = cv.getRotationMatrix2D((w/2, h/2), 45, 0.5)
dst = img //4
```

```python
affine_img = cv.warpAffine(img, mat, (w, h), borderValue=(0, 128, 255))
"""
cv.imshow("affine", affine_img)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(affine_img, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
affine_transparent = cv.warpAffine(img, mat, (w, h), borderMode=cv.BORDER_TRANSPARENT, dst=dst)
"""
cv.imshow("affine", affine_transparent)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(affine_transparent, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
affine_replicate = cv.warpAffine(img, mat, (w, h), borderMode=cv.BORDER_REPLICATE)
"""
cv.imshow("affine", affine_replicate)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(affine_replicate, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
affine_wrap = cv.warpAffine(img, mat, (w, h), borderMode=cv.BORDER_WRAP)
"""
cv.imshow("affine", affine_wrap)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(affine_wrap, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```
