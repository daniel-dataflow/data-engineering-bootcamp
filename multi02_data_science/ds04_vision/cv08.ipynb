```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../images/logo.png")

img = cv.resize(img, dsize=(0,0), fx=3, fy=3)

img_plot = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plot)
```

```python
# 내용 색상이 수축
kernel = np.ones((5, 5), np.uint8)
erode = cv.erode(img, kernel, iterations=2)

"""
cv.imshow("erode", erode)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(erode, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
# 내용 색상이 팽창
kernel = np.ones((5, 5), np.uint8)
dilates = cv.dilate(img, kernel, iterations=2)

"""
cv.imshow("dilates", dilates)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(dilates, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```

```python
"""
cv.MORPH_CLOSE      : Dilation(팽창) 후 Erosion(수축) 연산
cv.MORPH_OPEN       : Erosion 후 Dilation 연산
cv.MORPH_BLACKHAT   : Closing 과 원본의 차이 연산
cv.MORPH_GRADIENT   : Dilation과 Erosion의 차이 연산
cv.MORPH_TOPHAT     : Opening 과 원본의 차이 연산
"""

kernel = np.ones((5, 5), np.uint8)
ret, img_th = cv.threshold(img, 100, 255, cv.THRESH_BINARY)
gradient = cv.morphologyEx(img_th, cv.MORPH_GRADIENT, kernel)

result = cv.vconcat([img, gradient])

"""
cv.imshow("gradient", result)

cv.waitKey()
cv.destroyAllWindows()
"""
img_cv = cv.cvtColor(result, cv.COLOR_BGR2RGB)
plt.imshow(img_cv)
```
