```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("../images/headache.png")

img_plot = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_plot)
```

```python
# 임계값 처리
# cv.threshold(대상 이미지, 사용자 지정 임계값, 임계값보다 큰 값, 적용 타입)
# 100보다 작으면 0 100보다 크면 255로 하겠다.

th, img_th = cv.threshold(img, 100, 255, cv.THRESH_BINARY)
```

```python
'''
cv.imshow("threshold", img_th)

cv.waitKey()
cv.destroyAllWindows()
'''

plt.imshow(img_th)
```

```python
# 오츄 (OTSH) : 단일 채널에서만 가능!!! => 숙제
th, img_th = cv.threshold(img[:, :, 2], 0, 255, cv.THRESH_OTSU)
print(f"오츄 알고리즘에서의 최적임계값 : {th}")
```

```python
cv.imshow("threshold-otsu", img_th)

cv.waitKey(1)
cv.destroyAllWindows()
```
