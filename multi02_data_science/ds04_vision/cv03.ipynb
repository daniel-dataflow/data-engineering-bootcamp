```python
# 각채널의 값으로 히스토그램으로 체크해보자
# 맨 끝에 가장 많이 몰려 있는 그림이다.
plt.figure(figsize=(15, 5))

blue = cv.calcHist([img], [0], None, [256], [0, 256])
green = cv.calcHist([img], [1], None, [256], [0, 256])
red = cv.calcHist([img], [2], None, [256], [0, 256])

plt.subplot(131)
plt.plot(blue, color="b")

plt.subplot(132)
plt.plot(green, color="g")

plt.subplot(133)
plt.plot(red, color="r")
```
